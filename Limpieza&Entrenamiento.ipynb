```python
#ver si hay datos faltantes. Crear un procedimiento para ver si el dataset tiene datos faltantes.
def vereficarDatosFaltantes(df):
  for col in df.columns.tolist():
    print('{} missing values: {}'.format(col,df[col].isna().sum()),'.(',100*(df[col].isna().sum()/df.shape[0]) ,')') #Shape[0] cant filas y [1] cant columns

vereficarDatosFaltantes(df)
```

```python
df.describe()
```

```python
def EliminarDatosFaltantes(df,col):
  df=df[df[col].notna()]
  return df

df2=EliminarDatosFaltantes(df,"Name")
vereficarDatosFaltantes(df2)
```

```python
#print(df2["Age"].median())
#print(df2["Age"].mean())
#print(df2["Age"].mode())
df2["Age_Median"]=df2["Age"]
df2["Age_Mean"] =df2["Age"]
df2["Age_Median"].fillna(df2["Age"].median(),inplace=True)
df2["Age_Mean"].fillna(df2["Age"].mean(),inplace=True)
sns.displot([df["Age"],df2['Age_Median'],df2["Age_Mean"]],kde="True")
```

```python
def ReemplazarDatosFaltantes(df,col,medida):
  if(medida=="median"):
      df[col].fillna(df[col].media(),inplace=True)
  elif(medida=="mean"):
      df[col].fillna(df[col].mean(),inplace=True)
  else:
    df[col].fillna(df[col].mode(),inplace=True)
  return df
```

```python
from operator import irshift
def CuantificaOutliers(dataset):
  for col in dataset.columns:
    q1,q3=np.percentile(dataset[col],[25,75])
    iqr=q3-q1
    lower_bound=q1-(1.5*iqr)  #valores an√≥malos
    upper_bound=q3+(1.5*iqr)

    outlier = dataset[(dataset[col]< lower_bound)|(dataset[col]>upper_bound)]
    print(col,"Registros: ",outlier.shape[0],".Porc: ",outlier.shape[0]/dataset.shape[0]*100)

CuantificaOutliers(df2)
```
