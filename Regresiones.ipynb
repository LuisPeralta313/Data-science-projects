```python
import numpy as np
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import *
from sklearn.preprocessing import *
from sklearn.svm import *
```

```python
df = pd.read_csv('penguins_size.csv')
df.head()
```

# Verificar los valores unicos de las variables categoricas


```python
print(df['species'].unique())
print(df['island'].unique())
print(df['sex'].unique())
```

```python
# se trataran las variables categoricas codificando la frecuencia en la que aparecen los datos
# para asignarle un peso de cuantas veces se repiten los mismos
def FrequencyEncoding(dataset,columna):
  frequency = dataset[columna].value_counts(normalize =True)
  dataset[columna] = dataset[columna].map(frequency)
  return dataset[columna]
dataset_encoding = df
dataset_encoding['island'] = FrequencyEncoding(dataset_encoding,'island')
dataset_encoding['sex'] = FrequencyEncoding(dataset_encoding,'sex')
```

```python
# para las especies si utilizara
# Define el diccionario de mapeo de las especies a los valores num√©ricos
species_mapping = {'Adelie': 1, 'Chinstrap': 2, 'Gentoo': 3}

# Reemplaza los valores en la columna 'species' usando el diccionario de mapeo
dataset_encoding['species'] = dataset_encoding['species'].replace(species_mapping)
dataset_encoding.head()
```

# Verificar datos faltanes

```python
def display_missing_value(df):
    for col in df.columns.tolist():
        print('{} column missing values: {}'.format(col, df[col].isnull().sum()),'. (',(df[col].isnull().sum()/df.shape[0])*100,'%)')
    print('\n')
display_missing_value(dataset_encoding)
```

```python
# se trataran los datos faltantes utilizando la media de los datos
df_filled_mean = dataset_encoding.fillna(df.mean())
df_filled_mean.head()
```

# Analisis de correlacion

```python
corr = df_filled_mean.corr()
round(corr,2)
#eliminar  sex
```

```python
#eliminando las variables
df_filled_mean = df_filled_mean.drop(columns = ['sex'])
df_filled_mean.head()
```

# entrenar el modelo


```python
#seleccionar las variables
data1 = df_filled_mean
X = df_filled_mean.drop(["species"],axis=1)
Y = df_filled_mean["species"]
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=25,test_size=0.2)
```

# Regresion Lineal


```python
#evaluacion del modelo
lr = LinearRegression()
lr.fit(x_train, y_train)
#predecir el modelo
y_hat = lr.predict(x_test)
#Evaluar el modelo
print("Mean Squared Error ", mean_squared_error(y_test,y_hat))
print("Mean Absolute Error ",mean_absolute_error(y_test, y_hat))
print("R2 Score ",r2_score(y_test,y_hat))
```

# SVM

```python
svmodel = SVC()
#entrenamento del modelo

svmodel.fit(x_train,y_train)
#probando el modelo
y_hat = svmodel.predict(x_test)
#evaluando el modelo
print("Mean Squared Error ", mean_squared_error(y_test,y_hat))
print("Mean Absolute Error ",mean_absolute_error(y_test, y_hat))
print("R2 Score ",r2_score(y_test,y_hat))
```

# Regresion Polinomial

```python
polyfit = PolynomialFeatures(degree=2)
x_train_poly = polyfit.fit_transform(x_train)
x_test_poly = polyfit.fit_transform(x_test)
```

```python
#Entrenar al modelo
lrpoly = LinearRegression()
lrpoly.fit(x_train_poly, y_train)
#predecir el modelo
y_hat_poly = lrpoly.predict(x_test_poly)
#Evaluar el modelo

print("Mean Squared Error ", mean_squared_error(y_test,y_hat_poly))
print("Mean Absolute Error ",mean_absolute_error(y_test, y_hat_poly))
print("R2 Score ",r2_score(y_test,y_hat_poly))
```
